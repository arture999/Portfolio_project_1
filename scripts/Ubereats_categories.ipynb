```python
from urllib.request import Request, urlopen
from urllib.parse import quote_plus
from tqdm.notebook import tqdm
from bs4 import BeautifulSoup
import pandas as pd

import csv



HEADERS = {'User-Agent': 'Mozilla/5.0', 'Accept': 'text/html,application/xhtml+xml,application/xml'}
```

remove from my text file all the words food,except "fast food"

```python
# Reading categories from the file
with open(r"C:\Users\Arturo\anaconda3\envs\project1\mis archivos\categorias.txt", 'r') as f:
    categories = f.read().split('\n')

# Removing empty strings from the list
categories = [cat.strip() for cat in categories if cat.strip()]

# City URL
city_url = "https://www.ubereats.com/mx-en/category/mexico-city-df/"

# List to store final URLs
url_list = []

# Creating URLs with categories
for category in tqdm(categories, desc="Generating URLs"):
    # Replace spaces with hyphens and append to the city URL
    category_url = quote_plus(category.lower().replace(" ", "-"))
    full_url = f"{city_url}{category_url}"
    
    # Append the URL to the list
    url_list.append(full_url)

# Print the list of URLs
for url in url_list:
    print(url)
```

```python

### BUENO

# Assuming you have the url_list from the previous code
urls_set = set()

for url in tqdm(url_list, desc="Scraping URLs"):
    try:
        req = Request(url, headers=HEADERS)
        webpage = urlopen(req).read()
        soup = BeautifulSoup(webpage, 'html.parser')

        # Extract restaurant links from the current soup object
        for restaurant_div in soup.find_all("div", class_="ak c3"):
            for restaurant_link in restaurant_div.find_all("a", href=True):
                link = restaurant_link['href']
                urls_set.add(link)
                
    except Exception as e:
        print(f"Error scraping URL {url}: {e}")

# Convert the set to a DataFrame
df = pd.DataFrame(urls_set, columns=['url-restaurants'])

# Print or do further processing with the DataFrame
df.head()
```

```python
df.to_csv('restaurants_link_Mexico_city.csv', index=None)
```

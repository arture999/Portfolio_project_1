# WEB SCRAPPING

Scrapping the website UberEats only for learning purposes.

The data retrieved is the cities and states where Ubereats operates in Mexico.


```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import csv
```

```python
url = "https://www.ubereats.com/mx-en/location?slr=1"

response = requests.get(url)

soup = BeautifulSoup(response.text, 'html.parser')
```

```python
df = pd.DataFrame(columns=['city', 'state', 'url'])
```

```python
states = soup.find_all("div", class_="al bm b9")
states = [state for state in states if state.text != "All Countries"]

data = []

for state in states:
    print(f"Processing: {state.text}")

    city_container = state.find_next("div", class_="f8 f9 fa fk fc fd fe bb")

    if city_container:
        cities = city_container.find_all("a")

        for city in cities:
            city_name = city.text
            state_name = state.text
            city_url = "https://www.ubereats.com" + city['href']

            
            data.append({'city': city_name, 'state': state_name, 'url': city_url})

df = pd.concat([df, pd.DataFrame(data)], ignore_index=True)

print(df)
```
